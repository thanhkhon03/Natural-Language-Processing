# Biểu diễn các câu dưới dạng vec-to dựa trên TF-DF¶


```python
from sklearn.feature_extraction.text import TfidfVectorizer
sentences = [
    "Pytorch là framework được phát triển bởi Facebook.",
    "Đây là một ông lớn về công nghệ đầu tư rất nhiều nguồn lực cho việc phát triển Trí tuệ nhân tạo.",
    "Pytorch được phát triển với giấy phép mã nguồn mở do đó nó tạo được cho mình một cộng đồng rất lớn.",
    "Một cộng đồng lớn đồng nghĩa với nhiều tài nguyên để học và các vấn đề của bạn có thể đã có ai đó giải quyết và chia sẻ với cộng đồng.",
    "Pytorch cùng với Tensorflow và Keras là một trong những framework phổ biến được sử dụng trong các bài toán về Deep Learning hiện nay.",
    "Đặc biệt, trong các lĩnh vực nghiên cứu, hầu như các tác giả đều sử dụng pytorch để triển khai bài toán của mình.",
    "Pytorch cho thấy lợi thế của nó trong lĩnh vực nghiên cứu bởi việc rất dễ dàng để bạn debug và visualize, ngoài ra nó theo cơ chế Dynamic Graphs cho phép giảm thời gian huấn luyện mô hình."
]

#Khởi tạo vector
vectorizer = TfidfVectorizer()


#Ghép và biến đổi câu
tfidf_matrix = vectorizer.fit_transform(sentences)

#Lấy từ 
feature_names = vectorizer.get_feature_names_out()

#In vectơ TF-IDF cho mỗi câu
for i, sentence in enumerate(sentences):
    print(f"Câu {i+1}:")
    for j, word in enumerate(feature_names):
        if tfidf_matrix[i, j] > 0:
            print(f"    {word}: {tfidf_matrix[i, j]}")
```

# Biểu diễn các văn bản dưới dạng vec-to dựa trên TF-DF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Given text
given_text = "Pytorch là framework được phát triển bởi Facebook. Đây là một ông lớn về công nghệ đầu tư rất nhiều nguồn lực cho việc phát triển Trí tuệ nhân tạo. Pytorch được phát triển với giấy phép mã nguồn mở do đó nó tạo được cho mình một cộng đồng rất lớn.Một cộng đồng lớn đồng nghĩa với nhiều tài nguyên để học và các vấn đề của bạn có thể đã có ai đó giải quyết và chia sẻ với cộng đồng. Pytorch cùng với Tensorflow và Keras là một trong những framework phổ biến được sử dụng trong các bài toán về Deep Learning hiện nay. Đặc biệt, trong các lĩnh vực nghiên cứu, hầu như các tác giả đều sử dụng pytorch để triển khai bài toán của mình. Pytorch cho thấy lợi thế của nó trong lĩnh vực nghiên cứu bởi việc rất dễ dàng để bạn debug và visualize, ngoài ra nó theo cơ chế Dynamic Graphs cho phép giảm thời gian huấn luyện mô hình."

# Khởi tạo
vectorizer = TfidfVectorizer()

# Chuyển đổi văn bản
tfidf_vector = vectorizer.fit_transform([given_text])

# Lấy từ
feature_names = vectorizer.get_feature_names_out()

# In vector
print("Vector TF-IDF cho văn bản đã cho:")
for i, word in enumerate(feature_names):
    if tfidf_vector[0, i] > 0:
        print(f"    {word}: {tfidf_vector[0, i]}")
```
